```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
import time
import os
from datetime import datetime
```

```python
""" 
Para que se inicialice el navegador, habría que indicar la ruta del driver de tu navegador y 
asegurarte de que este incluida en el PATH de las variables de entorno del sistema 
"""
driver = webdriver.Chrome("C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")

# Abrir la página web
driver.get("https://geoportalgasolineras.es/geoportal-instalaciones/DescargarFicheros")

# Hacer click en "Información actualizada de precios":
info_button = driver.find_element_by_id("icon_precios")
info_button.click()

# Seleccionar una opción del menú desplegable
select_tipo = Select(driver.find_element_by_id('comboTipoBusqueda'))
select_tipo.select_by_value('EESS')

select_ubicacion = Select(driver.find_element_by_id('comboUbicacion'))
select_ubicacion.select_by_value('3')

select_carburante = Select(driver.find_element_by_id('comboCarburante'))
select_carburante.select_by_value('1')

# Hacer clic en el botón de descarga
download_button = driver.find_element_by_id("linkDescargaprecios")
download_button.click()

# Esperar a que se complete la descarga
time.sleep(20)

# Cerrar el navegador
driver.quit()

# Cambiar nombre al archivo:
# Si la carpeta de descargas no es la misma en donde estas trabajando, se debe indicar la ruta completa del archivo
nombre_actual = r'C:\Users\Usuario\Downloads\precios.xls'
fecha_actual = datetime.now().strftime('%d-%m-%Y')
nuevo_nombre_gasolina = f'gasolina_95_E5_{fecha_actual}.xls'
os.rename(nombre_actual, nuevo_nombre_gasolina)
```
